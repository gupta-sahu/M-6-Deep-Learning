```python
# Tanh - Hyperboliv Tangent
def tanh(z):
    return np.tanh(z)

z=np.linspace(20,-10,10)
y=tanh(z)

plt.plot(z,y,label="Tanh",color='Green')
plt.title("Tanh Activation Function")
plt.xlabel("Input")
plt.ylabel("Outpu")
plt.legend()
plt.show()
```
